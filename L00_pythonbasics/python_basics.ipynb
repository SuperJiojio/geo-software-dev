```python
import logging

def test_logging(x):
    # The log messages aren't usually delivered
    # in a notebook, but they would be passed
    # in the background to a logfile or database
    # depending on how you configure the settings.
    message = None
    if type(x) is int:
        message = "{} is an integer.".format(x)
        logging.info("{} is an integer.".format(x))
    elif type(x) is str:
        message = "{} is a string.".format(x)
        logging.info("{} is a string.".format(x))
    else:
        message = "What is this?: {}".format(x)
        logging.debug("What is this?: {}".format(x))
    # This is just so that you can see the text,
    # but the package would also give it a level like
    # info, debug, warning, error, critical, or custom
    # levels to help you figure out what is important.
    print(message)
```

```python
test_logging("48.7")
```

```python
test_logging(48.7)
```

```python
test_logging(48)
```
